```python
df = pd.read_csv('data_ventas.csv', delimiter=';')
```

```python
df
```

```python
## RFM
# R --> Recencia --> Cuando ha sido la ultima vez que nos ha comprado
# F --> Frecuencia --> Numero de veces que nos compra
# M --> Monto --> Monto que ha comprado un cliente
```

```python
df.columns
```

```python
## Las columnas que nos importan
## 'INT_LLAVE_SMNA' --> la semana que compro -- la recencia
## 'VCH_CODI_PEDI' --> codigo del pedido -- Frecuencia
## 'DEC_TOTL_PAGO_ML' --> Monto de cada pedido -- monto
## 'VCH_CODI_CNS' --> Codigo de cada cliente
```

```python
df1 = df[['VCH_CODI_CNS','VCH_CODI_PEDI','INT_LLAVE_SMNA','DEC_TOTL_PAGO_ML']]
```

```python
df1
```

```python
df1[df1['VCH_CODI_CNS']==1123910619]
```

```python
len(df1['VCH_CODI_CNS'].unique())
```

```python
df1['INT_LLAVE_SMNA'].map(str).str[0:6]
```

```python
df1['CAMPANA'] = df1['INT_LLAVE_SMNA'].map(str).str[0:6]
df1['CAMPANA'] = df1['CAMPANA'].map(int)
```

```python
df1
```

```python
df_temp = df1.groupby('VCH_CODI_CNS').agg({'DEC_TOTL_PAGO_ML':'sum','VCH_CODI_PEDI':'count','CAMPANA':'max'}).reset_index()
```

```python
df_temp
```

```python
## Cambiar los nombres de las variables
df_temp.rename(columns = {'VCH_CODI_CNS':'CLIENTE','DEC_TOTL_PAGO_ML':'MONTO','VCH_CODI_PEDI':'FRECUENCIA'},inplace=True)
```

```python
df_temp
```

```python
# PRESENTE: 202203. Para ver la recencia restamos la fecha presente con la fecha de la ultima venta
df_temp['RECENCIA'] = 202203 - df_temp['CAMPANA']
```

```python
df_temp['RECENCIA'].value_counts()
```

```python
## vamos a reemplazar los valores de la variable recencia para que tengan sentido
# 202203 - 202203 = 0
# 202203 - 202202 = 1
# 202203 - 202201 = 2
# 202203 - 202113 = 90 --> 3
# 202203 - 202112 = 91 --> 4
```

```python
df_temp['RECENCIA'] = df_temp['RECENCIA'].replace({90:3,91:4,92:5,93:6,94:7,95:8,96:9,97:10,98:11,99:12,100:13,101:14,102:15})
```

```python
df_temp['RECENCIA'].value_counts()
```

```python
df_temp
```

```python
df_temp['FRECUENCIA'].hist()
```

```python
print(df_temp['FRECUENCIA'].min())
print(df_temp['FRECUENCIA'].quantile(0.25))
print(df_temp['FRECUENCIA'].quantile(0.5))
print(df_temp['FRECUENCIA'].quantile(0.75))
print(df_temp['FRECUENCIA'].max())
```

```python
## creamos los buckets.
df_temp['FRECUENCIA_SCORE'] = pd.cut(df_temp['FRECUENCIA'],bins = [0,
                                                                   df_temp['FRECUENCIA'].quantile(0.25),
                                                                   df_temp['FRECUENCIA'].quantile(0.50),
                                                                   df_temp['FRECUENCIA'].quantile(0.75),
                                                                   df_temp['FRECUENCIA'].max()],
                                     labels = [1,2,3,4]).astype('int')
```

```python
df_temp
```

```python
df_temp['MONTO_SCORE'] = pd.cut(df_temp['MONTO'],bins = [0,
                                                                   df_temp['MONTO'].quantile(0.25),
                                                                   df_temp['MONTO'].quantile(0.50),
                                                                   df_temp['MONTO'].quantile(0.75),
                                                                   df_temp['MONTO'].max()],
                                     labels = [1,2,3,4]).astype('int')
```

```python
df_temp['RECENCIA_SCORE'] = pd.cut(df_temp['RECENCIA'],bins = [-1,
                                                                   df_temp['RECENCIA'].quantile(0.5),
                                                                   df_temp['RECENCIA'].quantile(0.75),
                                                                   df_temp['RECENCIA'].quantile(0.90),
                                                                   df_temp['RECENCIA'].max()],
                                     labels = [4,3,2,1]).astype('int')
```

```python
df_temp
```

```python
df_temp['RFM'] = df_temp['FRECUENCIA_SCORE'] + df_temp['MONTO_SCORE'] + df_temp['RECENCIA_SCORE']
```

```python
df_temp
```

```python
df_temp.groupby('RFM').agg({'CLIENTE':'count','MONTO':'mean','FRECUENCIA':'mean','RECENCIA':'mean'})
```
